```python
# Merge
import pandas as pd

x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,4],"C":[2,5,7,9]})
z=pd.merge(x,y,on="A")        #merge korte gele ekta part same rakhte hobe
print(z)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,4],"C":[2,5,7,9]})
z=pd.merge(y,x,on="A")        #j variable age dibo oita age show korbe
print(z)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,5],"C":[2,5,7,9]})
z=pd.merge(x,y,on="A")        #common values er index porjonto print hobe
print(z)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,5],"C":[2,5,7,9]})
z=pd.merge(x,y,how="left")     #left means B priority beshi
z1=pd.merge(x,y,how="right")   #right means c priority beshi
z2=pd.merge(x,y,how="outer")   #outer means B&C priority beshi
z3=pd.merge(x,y,how="outer",indicator=True)
print(z)
print()
print(z1)
print()
print(z2)
print()
print(z3)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,5],"B":[2,5,7,9]})
z=pd.merge(x,y,left_index=True,right_index=True)
print(z)
```

```python
#Concat
a=pd.Series([1,5,10,15])
b=pd.Series([10,50,15,120])
c=pd.concat([a,b])    #list akare dite hobe
print(c)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"A":[1,2,3,5],"C":[2,5,7,9]})
# y1=pd.DataFrame({"A":[1,2,3],"C":[2,5]})
z=pd.concat([x,y])                            #dataframe k o concat kora jai
zz=pd.concat([x,y],axis=1)                    #row akare ,axis=0 ,row
# z1=pd.concat([x,y1],axis=1)
# z2=pd.concat([x,y1],axis=1,join="inner")
z3=pd.concat([x,y],axis=1,keys=["x","y"])     #columns a x,y
z4=pd.concat([x,y],axis=0,keys=["x","y"])     #row te x,y
print(z)
print()
print(zz)
# print()
# print(z1)
# print()
# print(z2)
print()
print(z3)
print()
print(z4)
```

```python
#Groupby
var=pd.DataFrame({"name":["a","b","b","b","c","c","a","b"],"cl":[1,2,3,4,5,6,7,8],"kl":[2,4,6,8,3,5,10,12]})
print(var)
print()
var1=var.groupby("name")
print(var1)
for x,y in var1:
    print(x,y)
```

```python
var2=var1.get_group("a")
print(var2)
```

```python
var3=var1.min()
var4=var1.max()
var5=var1.mean()
print(var3)
print()
print(var4)
print()
print(var5)
```

```python
#join

x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]})
y=pd.DataFrame({"C":[1,2,5],"D":[2,5,9]})
x.join(y)
```

```python
x=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,6,8]},index=["a","b","c","d"])
y=pd.DataFrame({"C":[1,2,5],"D":[2,5,9]},index=["p","o","i"])
x.join(y,how="outer",lsuffix="_12")
```

```python
#melt
x1=pd.DataFrame({"A":[1,2,3,4,5],"B":[2,4,6,8,9]})
var=pd.melt(x1)  #vertically print hobe
print(var)    
print()

pd.melt(x1,id_vars="B")     #B index hisabe kaj korbe
pd.melt(x1,var_name="sami",value_name="tami")   #variable & value er name change hoye jbe
```
