```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

x=[2,4,12,8,10]
y=[1,8,5,7,9]

z=pd.DataFrame({"sami":x,"tamim":y})
sns.lineplot(x="sami",y="tamim",data=z)

plt.show()



```

```python
x=sns.load_dataset("penguins")
x
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,style="sex",hue="sex", palette="Accent")    #colorbar
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,style="sex",hue="sex", palette="Accent",markers=["o","D"])    
plt.show()
```
