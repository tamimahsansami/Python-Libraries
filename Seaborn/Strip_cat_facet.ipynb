```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
```

```python
#Catplot

sns.catplot(x="size",y="tip",data=var,hue="sex",height=6)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",height=5,kind="violin")
plt.show()      #box', 'boxen', 'violin', 'bar', 'count', and 'point'.
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",height=5,kind="boxen")
plt.show()  
```

```python
#FacetGrid  (multiple)


var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
