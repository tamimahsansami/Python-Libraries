```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
b=np.linspace(1,20,20).reshape(4,5)
b
```

```python
y={"color":"b","fontsize":"12"}
sns.heatmap(b,annot=True,annot_kws=y)
```

```python
a=np.random.randint(1,50,40).reshape(4,10)
                        
a
```

```python
sns.heatmap(a,annot=True,linewidth=4)
```

```python
sns.heatmap(a,annot=True,linewidth=1,linecolor="b")
```

```python
sns.heatmap(a,cbar=False)
```

```python
sns.heatmap(a,vmin=0,vmax=100,cmap="PuOr")
```

```python
#violinplot


y=sns.load_dataset("tips")
y
```

```python
sns.violinplot(x="day",y="total_bill",data=y)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=y,hue="sex",split=True)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=y,hue="sex",inner="stick")   #box', 'quart', 'stick', 'point', None],
plt.show()
```

```python
#Boxplot
y1=sns.load_dataset("tips")
y1
```

```python
sns.set(style="whitegrid")

s=sns.boxplot(x="day",y="total_bill",data=y1)
s
plt.show()
```

```python
s=sns.boxplot(x="day",y="total_bill",data=y1,hue="sex",palette='plasma')
s
plt.show()
```

```python
s=sns.boxplot(y="day",x="total_bill",data=y1,hue="sex",palette='plasma')
s
plt.show()
```
