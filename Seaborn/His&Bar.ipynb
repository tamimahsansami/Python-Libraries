```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
x=sns.load_dataset("penguins")
x
```

```python
sns.displot(x["bill_length_mm"])
plt.show()
```

```python
sns.displot(x["bill_length_mm"],bins=[30,35,40,45,50,55,60])
plt.show()
```

```python
sns.displot(x["bill_length_mm"],bins=[30,35,40,45,50,55,60],kde=True)
plt.show()
```

```python
sns.displot(x["bill_length_mm"],bins=[30,35,40,45,50,55,60],kde=True,rug=True)
plt.show()
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island,y=var.flipper_length_mm)     #labels are pre defined
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var)     #labels are pre defined
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex")     
plt.show()
```

```python
or_1=['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1)     
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,hue_order=['Female','Male'])     
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,hue_order=['Female','Male'],n_boot=40)     
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,hue_order=['Female','Male'],palette="icefire")     
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,hue_order=['Female','Male'],saturation=0.3)     
plt.show()
```

```python
sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,
            hue_order=['Female','Male'],errorbar=('ci', 100),err_kws={'color': 'r'},capsize=.1)     
plt.show()
```

```python
sns.set(style="darkgrid")

sns.barplot(x='island',y='flipper_length_mm',data=var,hue="sex",order=or_1,hue_order=['Female','Male'],palette="icefire")     
plt.show()
```
