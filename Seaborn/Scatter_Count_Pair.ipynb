```python
#Countplot

var=sns.load_dataset("penguins")
var
```

```python


sns.countplot(x="sex",data=var,hue="sex")
plt.show()
```

```python
#pairplot

var=sns.load_dataset("penguins")
var
```

```python
sns.pairplot(var)    #numerical values
plt.show()
```

```python
sns.pairplot(var,hue="sex")    
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")    
plt.show()
```

```python
sns.pairplot(var,hue="sex",kind="kde")    #scatter, kde, hist, reg
plt.show()
```

```python
sns.pairplot(var,hue="sex",kind="kde",diag_kind="hist")    
plt.show()
```
