```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,3,5,7]
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.pie(x=[1])

x2=[1,2,3,4]
plt.subplot(2,2,3)
plt.pie(x2)

x3=["a","b","c","d"]
x4=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x3,x4)


plt.savefig("Sub",dpi=2000)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,3,5,7]
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.pie(x=[1])

x2=[1,2,3,4]
plt.subplot(2,2,3)
plt.pie(x2)

x3=["a","b","c","d"]
x4=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x3,x4)


plt.savefig("Sub",dpi=1000,facecolor="b")
plt.show()
```

```python
plt.savefig("Sub.pdf",dpi=1000)
plt.show()
```

```python
plt.savefig("Sub2",dpi=1000,facecolor="y",bbox_inches="tight")
plt.show()
```

```python
x=[10,20,30,40,50,60,90]
y=[12,40,50,10,70,30,50]
x1=[12,14,35,45,47,50,80]
y1=[45,67,34,23,15,57,30]
l=['Area1','Area2','Area3']

plt.stackplot(x,y,x1,y1,labels=l)
plt.legend(loc=2)

plt.savefig("Sub3",dpi=1000,facecolor="y",bbox_inches="tight")
plt.show()
```

```python
x=[10,25,35,40,50,60,90]
y=[15,25,35,45,85,65,75]

plt.xticks(x,labels=['A','B','C','D','F','G',"H"])
plt.yticks(y)

plt.plot(x,y)
plt.show()
```

```python
plt.xticks(x)
plt.yticks(y)
plt.xlim(0,150)
plt.ylim(0,200)

plt.plot(x,y)
plt.show()
```
