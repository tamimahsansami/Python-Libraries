```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,4,6,8,10]
y=[1,3,5,7,9,11]

c=["r","m","b","g","y","r"]
s=[20,30,400,35,60,80]
plt.xlabel("Plot",fontsize=10)
plt.ylabel("Bar",fontsize=10)
plt.title("Funny")

plt.scatter(x,y,color=c,sizes=s)     #color    sizes
```

```python
x=[1,2,4,6,8,10]
y=[1,3,5,7,9,11]

c=[0.4,20,30,400,50,40]       #uses colormaps
s=[200,300,400,350,60,800]
plt.xlabel("Plot",fontsize=10)
plt.ylabel("Bar",fontsize=10)
plt.title("Funny")

plt.scatter(x,y,c,sizes=s,alpha=.4,cmap="BrBG")
z=plt.colorbar()
z.set_label("colorbar")    #colorbar a name add korte
```
