```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show
```

```python
plt.boxplot(x,notch=True)
plt.show
```

```python
plt.boxplot(x,notch=True,vert=False)
plt.show
```

```python
plt.boxplot(x,widths=.3)
plt.show
```

```python
plt.boxplot(x,widths=.2,labels=["Python"],patch_artist=True)
plt.show
```

```python
plt.boxplot(x,labels=["Python"],widths=.3,showmeans=True)
plt.show
```

```python
x=[10,20,30,40,50,60,70,140]
plt.boxplot(x)
plt.show
```

```python
x=[10,20,30,40,50,60,70,140]
plt.boxplot(x,sym="g*")     #color & shape
plt.show
```

```python
x=[10,20,30,40,50,60,70,140]
plt.boxplot(x,whis=2.8)     #join with the outlayer
plt.show
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=["Python"],boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="g"))
plt.show
```

```python
x=[10,20,30,40,50,60,70]
y=[20,40,80,100,90,30,50]
z=[x,y]
plt.boxplot(z,labels=["Python","c++"],boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="g"))
plt.show
```

```python
x=[10,20,30,40,50,60,70]
y=[20,40,80,100,90,30,50]
x1=[12,14,35,45,47,50,80]
y1=[45,67,34,23,15,57,30]
z=[x,x1,y,y1]
plt.boxplot(z,labels=["Python","c++","Java","R"],boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="g"))
plt.show
```
