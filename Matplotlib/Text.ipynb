```python
import matplotlib.pyplot as plt
y=[2,4,6,8]
x=[1,2,3,4]

plt.xlabel("X AXIS",fontsize=15)
plt.ylabel("Y AXIS",fontsize=15)
plt.title("RANDOM FILE")
c=["r","y","r","g"]

plt.text(1,7,"SAmi",fontsize=15,style='italic')
plt.bar(x,y,width=.3,color=c)
```

```python
plt.text(1,7,"SAmi",fontsize=15,style='italic',bbox={"facecolor":"blue"})
plt.bar(x,y,width=.3,color=c)
```

```python

y=[2,4,6,8]
x=[1,2,3,4]

plt.xlabel("X AXIS",fontsize=15)
plt.ylabel("Y AXIS",fontsize=15)
plt.title("RANDOM FILE")
c=["r","y","r","g"]

plt.text(1,7,"SAmi",fontsize=15,style='italic',bbox={"facecolor":"blue"})

plt.annotate("Tamim",fontsize=15,xy=(2,0),xytext=(3,7),arrowprops=dict(facecolor="blue",shrink=100))

plt.bar(x,y,width=.3,color=c)
```
