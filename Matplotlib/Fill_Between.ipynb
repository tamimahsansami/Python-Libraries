```python
import matplotlib.pyplot as plt

x=[2,4,6,8,10]
y=[1,3,5,7,9]
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[2,4,6,8,10]
y=[1,3,5,7,9]
plt.plot(x,y)
plt.fill_between(x=[4,6],y1=3,y2=5)

plt.show()
```

```python
import numpy as np     #numpy works as a number
x=np.array([2,4,6,8,10])
y=np.array([1,3,5,7,9])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=4)&(x<=6))

plt.show()
```
