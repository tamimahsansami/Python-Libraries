```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,90]
y=[12,40,50,10,70,30,50]
x1=[12,14,35,45,47,50,80]
y1=[45,67,34,23,15,57,30]
l=['Area1','Area2','Area3']

plt.stackplot(x,y,x1,y1,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
plt.stackplot(x,y,x1,y1,labels=l,colors=["r","g","b"])
plt.legend(loc=2)
plt.show()
```

```python
plt.stackplot(x,y,x1,y1,labels=l,colors=["r","g","b"],baseline="sym")
plt.legend(loc=2)
plt.show()
```

```python
plt.stackplot(x,y,x1,y1,labels=l,colors=["r","g","b"])
plt.title("Python")
plt.legend(loc=2)
plt.show()
```
