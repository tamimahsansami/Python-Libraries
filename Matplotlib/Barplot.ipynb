```python
import matplotlib.pyplot as plt
y=[2,4,6,8]
x=[1,2,3,4]
plt.bar(x,y,width=.5)
plt.show()
```

```python
x=["numpy","pandas","matplotlib","seaborn"]
y=[2,4,6,8]
plt.bar(x,y,width=.5)
plt.show()
```

```python
plt.xlabel("X AXIS",fontsize=15)
plt.ylabel("Y AXIS",fontsize=15)
plt.title("RANDOM FILE")
c=["r","y","r","g"]
plt.bar(x,y,width=.3,color=c)

plt.show()
```

```python
plt.bar(x,y,width=.3,color=c,align="edge")     #xaxis a label change hobe
```

```python
plt.bar(x,y,width=.3,linewidth=.1,color=c,align="center")   #bydefault center a thake
```

```python
plt.bar(x,y,width=.3,color=c,edgecolor="r",linewidth=1.5)    #boundary color, linewidth=boundary width
```

```python
plt.bar(x,y,width=.3,color=c,edgecolor="r",linewidth=1.5,linestyle=":",alpha=.4)   #alpha diye bar light & dark korte hoi(0-1)
```

```python
x=["numpy","pandas","matplotlib","seaborn"]
y=[10,5,10,5]
z=[5,1,5,2]

plt.xlabel("X AXIS",fontsize=15)
plt.ylabel("Y AXIS",fontsize=15)
plt.title("RANDOM FILE")

plt.bar(x,y,width=.3,color="b",label="sami")
plt.bar(x,z,width=.3,color="r",label="tamim")
plt.legend()
plt.show()
```

```python
import numpy as np
x=["numpy","pandas","matplotlib","seaborn"]
y=[10,5,10,5]
z=[5,1,5,2]

width=.2
p=np.arange(len(x))    #x variable er index as array
q=[j+width for j in p]

plt.xlabel("X AXIS",fontsize=15)
plt.ylabel("Y AXIS",fontsize=15)
plt.title("RANDOM FILE")

plt.bar(q,y,width,color="b",label="sami")
plt.bar(q,z,width,color="r",label="tamim")     # plt.xticks(x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
company=["amazon","google","nasa","intel"]
pro=[35,20,30,40]
xpo=[12,15,20,35]

y=np.arange(len(company))


plt.xticks(y,company)     #values of company will be replaced by y
plt.bar(y-0.2,pro,color="b",width=.4,label="sami")
plt.bar(y+0.2,xpo,color="g",width=.4,label="tamim")
plt.legend()     #label er jonno
plt.show()
```
