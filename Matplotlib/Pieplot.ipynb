```python
import matplotlib.pyplot as plt

x=[20,40,60,80,50]
y=["Java","C++","C","Python","R"]

plt.pie(x,labels=y)
plt.show()
```

```python
x=[20,40,60,80,50]
y=["Java","C++","C","Python","R"]
ex=[0.1,0.1,0.2,0.1,0.1]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[20,40,60,80,50]
y=["Java","C++","C","Python","R"]
ex=[0.1,0.1,0.2,0.1,0.1]
c=['r','g','y','m','b']
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%")
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",labeldistance=.5)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",startangle=0,textprops={"fontsize":15})
plt.show()
```

```python
ex=[0.0,0.0,0.2,0.0,0.0]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,wedgeprops={"linewidth":5,"edgecolor":'k'})
plt.show()
```

```python
ex=[0.0,0.0,0.2,0.0,0.0]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,rotatelabels=True)
plt.title("SAmi")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1],radius=.8)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[20,40,60,80,50]
y=[15,25,35,60,45]
z=["Java","C++","C","Python","R"]

c=['r','g','y','b','m']
plt.pie(x,labels=z,radius=1.2,colors=c)
plt.pie(y,radius=.4)
plt.show()
```

```python
x=[20,40,60,80,50]
y=[15,25,35,60,45]
z=["Java","C++","C","Python","R"]

c=['r','g','y','b','m']
plt.pie(x,labels=z,radius=1.2,colors=c)
plt.pie([1],radius=.8,colors='w')
plt.show()
```
