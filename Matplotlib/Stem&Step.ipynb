```python
import matplotlib.pyplot as plt

y=[1,3,5,7]
z=[4,5,6,2]

plt.stem(y,z)
plt.show()
```

```python
plt.stem(y,z,linefmt="--")

plt.show()
```

```python
plt.stem(y,z,linefmt=":",markerfmt="D",basefmt="g")
plt.show()
```

```python
plt.stem(y,z,linefmt=":",markerfmt="D",basefmt="g",orientation="horizontal")
plt.show()
```

```python
import matplotlib.pyplot as plt

y=[1,3,5,7]
z=[1,3,6,8]

plt.step(y,z)
plt.grid()
plt.show()
```

```python
plt.step(y,z,marker="o",mfc="r",ms=15,label="random")
plt.legend(loc=2)      #1 means upperright,  2 =upperleft,  3=downleft   ,4=downright
plt.grid()
plt.show()
```
